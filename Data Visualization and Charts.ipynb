```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```


Life expectancy at birth indicates the number of years a newborn baby would live if the mortality patterns prevailing at the time of birth remained the same throughout its life. It is a key indicator for assessing population health.

For this work, we will first study the distribution of life expectancy in the world and compare life expectancy in different regions of the world; we will then make a comparative study of the factors affecting life expectancy in North America and, finally, we will study the evolution of life expectancy in the United States over time.



```python
#we read the dataset
df = pd.read_csv("life expectancy.csv")
df
```



we plot a histogram to see the distribution of life expectancy around the world 



```python
sns.histplot(x='Life Expectancy World Bank', data= df, kde= True)
```

```python
df['Life Expectancy World Bank'].describe()
```



We plot a barplot to compare life expectancy in different regions of the world. 



```python
sns.barplot(x="Life Expectancy World Bank",y= "Region",  data= df)
```


we note that life expectancy is highest in North America



Now, let's do a North American comparative study on life experience and the factors that influence life experience.


```python
df2 = df[df["Region"] == 'North America']
df2
```

```python
#let's try the histogram first
sns.histplot(x='Life Expectancy World Bank',  data= df2, hue = 'Country Name', kde= True)
```


we note that it is very difficult to compare the life expectancy of these three countries with the histogram, as the graph is very cluttered, making interpretation very complicated.



Next, let's compare with the Barplot


```python
sns.barplot(x="Country Name",y= "Life Expectancy World Bank",  data= df2)
```


This diagram shows that Canada has the highest life expectancy, followed by Bermuda, and that the USA has the lowest.




Let's go a step further and compare the life expectancy of these three countries using the box plot, then compare the median, maximum, and minimum of the three life expectancies.


```python
sns.boxplot(x ='Country Name', y='Life Expectancy World Bank', data= df2 )
```


we find that canada's maximum, minimum and median life expectancy are higher than those of other countries, while the u.s.'s maximum life expectancy is lower than canada's minimum.





Let's now compare the evolution of life expectancy in the three countries over the years.



```python
sns.lineplot(data= df2, x= "Year", y="Life Expectancy World Bank", hue = 'Country Name')
```

```python
sns.scatterplot(data= df2, x= "Year", y="Life Expectancy World Bank", hue = 'Country Name')
```


In all three countries, life expectancy is increasing over the years. and the lineplot provides a better interpretation than the scatterplot.

Among all these variables, let's check which one actually affects life expectancy. The study will be carried out in the United States, a country with the lowest life expectancy in North America.

```python
df1 = df[df["Country Code"] == 'USA']
df1
```

```python
df1['Life Expectancy World Bank'].describe()
```

```python
#we use pairplot to see the variation between different variables
sns.pairplot(df1)
```
